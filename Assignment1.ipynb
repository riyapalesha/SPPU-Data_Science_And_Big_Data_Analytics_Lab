**HANDLING MISSING VALUES**


There are several ways to handle records containing missing values (NaN) in a dataset. The choice of method depends on the specific context and the nature of the data. Here are some common approaches:

Dropping rows: Remove the rows that contain missing values. This approach is suitable when the missing values are limited and removing the affected rows does not significantly impact the analysis. You can use the dropna() method in Pandas with the subset parameter to specify the columns to consider when dropping rows.

Dropping columns: Remove entire columns that contain a large number of missing values or are irrelevant to the analysis. This approach is appropriate when the missing values are prevalent in specific columns and those columns do not provide meaningful information for the analysis. Use the dropna() method with the axis=1 parameter to drop columns.

Imputation: Replace missing values with estimated or calculated values. Imputation allows you to fill in the missing values based on various strategies. Common imputation techniques include replacing missing values with the mean, median, mode, or a constant value. The Pandas fillna() method can be used to fill missing values with a specific value or a calculated value.

get_dummies() method 2

the pd.get_dummies() function is used to convert the 'Category' column of the DataFrame df into multiple binary-encoded columns. Each unique category in the 'Category' column is represented by a separate column, where a value of 1 indicates the presence of that category in the original observation, and 0 indicates its absence.

The prefix parameter is used to add a prefix to the column names in the encoded DataFrame to distinguish them from the original column. In this case, the prefix 'Category' is added to the encoded columns.

```python
df_encoded=pd.get_dummies(df["Status"], prefix="Status")
df_encoded
```

**Data Normalization**
1. MinMax algorithm
2. Z score

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
column_values=df["BasePay"].values.reshape(-1,1)
normalized_values=scaler.fit_transform(column_values)
df["NormalizedBasePay"]=normalized_values
```

```python
df.tail()
```

```python
# Min Max algorithm using formula
op_min=df["OvertimePay"].min()
op_max=df["OvertimePay"].max()


df["NormalizedOvertimePay"]=(df["OvertimePay"]-op_min)/(op_max-op_min)
```

```python
df["OtherPay"].dtype
```

```python
# Z-score normalization technique->(x - mean) / std

otherPay_mean = df["OtherPay"].mean()
otherPay_std = df["OtherPay"].std()

df["NormalizedOtherPay"] = (df["OtherPay"] - otherPay_mean) / otherPay_std
```
