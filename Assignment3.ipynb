# Method 1 (By making age groups)


age_ranges = [0, 18, 26, 36, 46, 56, income_df['age'].max() + 1]: This line defines the age ranges, similar to the previous code. The age ranges are specified as [0, 18, 26, 36, 46, 56, max_age + 1], where max_age is the maximum age in the 'age' column of the DataFrame income_df.

age_labels = ['Below 18', '18-25', '26-35', '36-45', '46-55', '56+']: This line defines the labels for each age range, just like before.

sample = pd.cut(income_df['age'], bins=age_ranges, labels=age_labels, right=False): This line applies the pd.cut() function to categorize the 'age' variable into the specified age ranges, just as you did earlier. The resulting categorical variable is stored in the sample variable.

grouped_data = income_df.groupby(sample): This line groups the DataFrame income_df based on the values in the sample variable. This creates a DataFrameGroupBy object named grouped_data, which represents the grouped data.

summary_stats = grouped_data['income'].agg(['mean', 'median', 'min', 'max', 'std']): This line calculates the summary statistics for the 'income' variable within each group. The .agg() method is applied to the 'income' column of the grouped_data object, specifying the desired aggregation functions ['mean', 'median', 'min', 'max', 'std']. The resulting summary statistics are stored in the summary_stats DataFrame


```python
age_ranges = [0, 18, 26, 36, 46, 56, income_df['age'].max() + 1]
age_labels = ['Below 18', '18-25', '26-35', '36-45', '46-55', '56+']
```

```python
sample = pd.cut(income_df['age'], bins=age_ranges, labels=age_labels, right=False)
```

```python
sample
```

```python
grouped_data = income_df.groupby(sample)
for group_name, group_data in grouped_data:
    print(f"Age Group: {group_name}")
    print(group_data)
    print()
```

```python
summary_stats = grouped_data['income'].agg(['mean', 'median', 'min', 'max', 'std'])
```

```python
summary_stats
```
