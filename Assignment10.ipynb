#Outliers

Outliers are data points that deviate significantly from the majority of the data and can be useful for understanding data quality or uncovering anomalies.

#Boxplot

 box plot displays the distribution of each numeric variable in the Iris dataset (sepal length, sepal width, petal length, and petal width). The box plot represents the median, quartiles, and any potential outliers for each variable.

#Violinplot

 The violin plot includes a kernel density estimation (KDE) curve, which provides insights into the shape and spread of the data distribution.

#Kdeplot

 The KDE curve provides a smooth estimate of the distribution, showing the shape and concentration of data points across different values.


```python
sepal_width=df["SepalWidthCm"]
Q1=sepal_width.quantile(0.25)
Q3=sepal_width.quantile(0.75)
print("Q1 = ",Q1)
print("Q3 = ",Q3)
IQR=Q3-Q1
print("IQR = ",IQR)
lowerbound=Q1-1.5*IQR
upperbound=Q3+1.5*IQR
print("Lower bound = ",lowerbound)
print("Upper bound = ",upperbound)
```

```python
outliers=sepal_width[(sepal_width<lowerbound) | (sepal_width>upperbound)]
print("Outliers in SepalWidthCm ")
print(outliers)
```
